Q4) The population of a school is divided into 60% boys and 40% girls. If you want equal representation
of both genders in the sample, which method should you use: Simple Random Sampling or Stratified Sampling?Why?

ans) School population: 60% boys, 40% girls

Goal: Equal representation of boys and girls in the sample:

SIMPLE RANDOM SAMPLING  (SRS):


In SRS, every student has an equal chance of being selected.

If you draw a sample randomly, the proportion of boys and girls in the sample will likely reflect the population (â‰ˆ 60% boys, 40% girls).

This means unequal representation â€” boys will dominate the sample.

STRATIFIED SAMPLING:

In stratified sampling, the population is divided into strata (here: boys and girls).

You then select samples deliberately from each stratum.

If you want equal representation, you can take the same number of boys and girls regardless of their population proportion.

This ensures balanced representation of both genders in the sample.

Final Answer:

You should use Stratified Sampling, because it allows you to control the proportion of boys and girls in the sample. Simple Random Sampling would mirror the population ratio (60:40), but Stratified Sampling ensures equal representation as required.

Q10) Differentiate between discrete and continuous random variables with examples.

ans) A random variable is a numerical outcome of a random phenomenon, and it can be classified as either discrete or continuous depending on the type of values it can take.

A discrete random variable assumes distinct, countable values, often arising from situations where outcomes can be listed individually.

For example, the number of heads obtained when tossing three coins or the number of students absent in a class are discrete random variables because they can only take whole number values like 0, 1, 2, and so on.

In contrast, a continuous random variable can take any value within a given interval, representing measurements that vary smoothly without gaps.

For instance, the height of students in a school or the time taken to complete a task are continuous random variables because they can assume infinitely many possible values within a range, such as 160.5 cm or 12.37 minutes.

In summary, discrete random variables deal with countable outcomes, while continuous random variables deal with measurable outcomes that can take on any value within an interval.
